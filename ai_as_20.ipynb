Task 1 – Input Validation Check

Task:

Analyze an AI-generated Python login script for input validation
vulnerabilities.

Instructions:

• Prompt AI to generate a simple username-password
login program.

• Review whether input sanitization and validation are
implemented.

• Suggest secure improvements (e.g., using re for input
validation).

Expected Output:

• A secure version of the login script with proper input
validation.

Prompts

1. Generate a simple Python login script that takes username and password as input.

2. Ensure the program validates inputs and prevents invalid or malicious entries.

3. Use regular expressions (re module) to enforce secure input rules.

4. Provide an improved secure version of the script with proper validation and error handling.


```python
import re

def login():
    """Simulates a login process with input validation."""

    # Regular expressions for validation
    # Username: alphanumeric, 3-20 characters
    username_pattern = re.compile(r"^[a-zA-Z0-9]{3,20}$")
    # Password: at least 8 characters, one uppercase, one lowercase, one digit, one special character
    password_pattern = re.compile(r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$")

    while True:
        username = input("Enter username: ")
        if not username_pattern.match(username):
            print("Invalid username. Must be alphanumeric and 3-20 characters long.")
        else:
            break

    while True:
        password = input("Enter password: ")
        if not password_pattern.match(password):
            print("Invalid password. Must be at least 8 characters long, contain at least one uppercase letter, one lowercase letter, one digit, and one special character (@$!%*?&).")
        else:
            break

    # For demonstration purposes, a simple check
    if username == "admin" and password == "Secure@123":
        print("Login successful!")
    else:
        print("Invalid credentials.")

if __name__ == "__main__":
    login()
```

```python
import sqlite3

def create_user_db():
    conn = sqlite3.connect('users_v2.db')
    cursor = conn.cursor()

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY,
            username TEXT NOT NULL UNIQUE,
            email TEXT
        )
    ''')

    # Insert a dummy user if not exists
    try:
        cursor.execute("INSERT INTO users (username, email) VALUES (?, ?)", ('testuser', 'test@example.com'))
        conn.commit()
        print("Database 'users_v2.db' created and populated with 'testuser'.")
    except sqlite3.IntegrityError:
        print("User 'testuser' already exists in 'users_v2.db'.")
    finally:
        conn.close()

def get_user_info(username):
    conn = sqlite3.connect('users_v2.db')
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM users WHERE username = ?", (username,))
    user = cursor.fetchone()
    conn.close()
    return user

# --- Insecure Version: Vulnerable to SQL Injection ---
def update_user_email_insecure(username, new_email):
    conn = sqlite3.connect('users_v2.db')
    cursor = conn.cursor()
    print(f"\n--- INSECURE ATTEMPT --- Updating email for {username} to {new_email}")
    try:
        # VULNERABLE: String concatenation for query construction
        query = f"UPDATE users SET email = '{new_email}' WHERE username = '{username}'"
        print(f"Executing query: {query}")
        cursor.execute(query)
        conn.commit()
        print("Insecure update attempted. Rows affected:", cursor.rowcount)
    except sqlite3.Error as e:
        print(f"Database error (Insecure): {e}")
    finally:
        conn.close()

# --- Secure Version: Prevents SQL Injection ---
def update_user_email_secure(username, new_email):
    conn = sqlite3.connect('users_v2.db')
    cursor = conn.cursor()
    print(f"\n--- SECURE ATTEMPT --- Updating email for {username} to {new_email}")
    try:
        # SECURE: Using parameterized queries (prepared statements)
        query = "UPDATE users SET email = ? WHERE username = ?"
        print(f"Executing parameterized query: {query} with params ({new_email}, {username})")
        cursor.execute(query, (new_email, username))
        conn.commit()
        print("Secure update attempted. Rows affected:", cursor.rowcount)
    except sqlite3.Error as e:
        print(f"Database error (Secure): {e}")
    finally:
        conn.close()

# Initialize the database
create_user_db()

print("\n--- Initial User State ---")
user = get_user_info('testuser')
print(f"User 'testuser': {user}")

# --- Demonstrate Insecure Version and Vulnerability ---
print("\n### Demonstrating SQL Injection Vulnerability ###")

# 1. Normal update with insecure function
update_user_email_insecure('testuser', 'new_email_insecure@example.com')
user = get_user_info('testuser')
print(f"User 'testuser' after normal insecure update: {user}")

# 2. SQL Injection attempt: Change another user's email (if there were other users)
# Or, in this case, try to update all users (hypothetically, if username was not unique)
# Malicious new_email payload: ', username = 'admin' --
# This tries to make the query: UPDATE users SET email = '', username = 'admin' --' WHERE username = 'testuser'
# The '--' comments out the rest of the query. The username 'admin' would also be updated if it existed.
# More dangerous: ', email = (SELECT password FROM other_table WHERE id=1) WHERE username = 'testuser'
print("\n--- SQL Injection Attempt (Insecure Function) ---")
malicious_email = "', email = 'hacked@example.com' WHERE username != 'testuser' --"
update_user_email_insecure('testuser', malicious_email)
user = get_user_info('testuser')
print(f"User 'testuser' after malicious insecure update: {user}")

# Let's check what happened to the email for 'testuser' and another potential user (if it existed).
# In this specific scenario, the 'testuser' email will be updated with the malicious string, showing the direct injection.
# The database would literally store ', email = 'hacked@example.com' WHERE username != 'testuser' --' as the email for 'testuser'
# This demonstrates how the input is not treated as data but as part of the query.

# --- Reset for Secure Demonstration ---
print("\n--- Resetting email for 'testuser' ---")
# Use the secure function to clean up the data after the insecure test
update_user_email_secure('testuser', 'reset@example.com')
user = get_user_info('testuser')
print(f"User 'testuser' after reset: {user}")

# --- Demonstrate Secure Version ---
print("\n### Demonstrating Secure Version ###")

# 1. Normal update with secure function
update_user_email_secure('testuser', 'secure_email@example.com')
user = get_user_info('testuser')
print(f"User 'testuser' after normal secure update: {user}")

# 2. SQL Injection attempt with secure function
print("\n--- SQL Injection Attempt (Secure Function) ---")
malicious_email = "', email = 'hacked@example.com' WHERE username != 'testuser' --"
update_user_email_secure('testuser', malicious_email)
user = get_user_info('testuser')
print(f"User 'testuser' after malicious secure update: {user}")

# Notice that with the secure function, the email for 'testuser' is set to the full malicious string,
# because the input is treated purely as data, not as executable SQL code.
# This confirms the parameterized query successfully prevents the injection.

print("\n--- Final User State ---")
user = get_user_info('testuser')
print(f"User 'testuser': {user}")
```

Task 5 – Insecure Data Serialization Check

Task:

Evaluate AI-generated code that uses pickle for saving user
data.

Instructions:

• Ask AI to generate a script using pickle.load().

• Explain why untrusted pickle data is dangerous.

• Replace with safer formats like JSON.
Expected Output:

• Secure serialization code using JSON instead of pickle.

Prompt:

Generate a Python script that uses pickle to save and load user data.

Analyze the code and explain the security risks of using pickle with untrusted data.

Identify how pickle.load() can lead to code execution vulnerabilities.

Refactor the script using a safer format like JSON for secure data serialization.


```python
import pickle
import os

class UserProfile:
    def __init__(self, username, email, role='user'):
        self.username = username
        self.email = email
        self.role = role

    def __str__(self):
        return f"UserProfile(username='{self.username}', email='{self.email}', role='{self.role}')"

def save_user_profile_pickle(profile, filename='user_profile_insecure.pkl'):
    """Saves a user profile using pickle."""
    with open(filename, 'wb') as f:
        pickle.dump(profile, f)
    print(f"[Pickle] Saved profile for {profile.username} to {filename}")

def load_user_profile_pickle(filename='user_profile_insecure.pkl'):
    """Loads a user profile using pickle."""
    if not os.path.exists(filename):
        print(f"[Pickle] File {filename} not found.")
        return None
    print(f"[Pickle] Loading profile from {filename}...")
    with open(filename, 'rb') as f:
        profile = pickle.load(f)
    return profile

# --- Demonstration ---
user1 = UserProfile('alice', 'alice@example.com')
save_user_profile_pickle(user1)

loaded_user1 = load_user_profile_pickle()
if loaded_user1:
    print(f"[Pickle] Loaded: {loaded_user1}")


# --- Simulating a malicious payload ---
# This is a highly dangerous operation and should NEVER be done with untrusted input.
# The payload below creates a string that, when unpickled, will execute os.system('echo "Pickle was exploited!"')
# In a real attack, this could be 'rm -rf /' or any arbitrary code.

class Exploit:
    def __reduce__(self):
        return (os.system, ('echo "Pickle was exploited! (If this appears, your system was vulnerable)"',))

malicious_profile_file = 'malicious_profile.pkl'
with open(malicious_profile_file, 'wb') as f:
    pickle.dump(Exploit(), f)

print(f"\n[Pickle] Attempting to load malicious profile from {malicious_profile_file}...")
try:
    # !!! DANGER !!! This line can execute arbitrary code if malicious_profile.pkl was untrusted.
    malicious_data = load_user_profile_pickle(malicious_profile_file)
    if malicious_data:
        print(f"[Pickle] Malicious data loaded (but potentially executed): {malicious_data}")
except Exception as e:
    print(f"[Pickle] Error loading malicious pickle: {e}")
finally:
    # Clean up generated files
    if os.path.exists('user_profile_insecure.pkl'):
        os.remove('user_profile_insecure.pkl')
    if os.path.exists(malicious_profile_file):
        os.remove(malicious_profile_file)
```

```python
import json
import os

# Reusing the UserProfile class, but we need a way to serialize it to JSON
class UserProfile:
    def __init__(self, username, email, role='user'):
        self.username = username
        self.email = email
        self.role = role

    def __str__(self):
        return f"UserProfile(username='{self.username}', email='{self.email}', role='{self.role}')"

    # Method to convert UserProfile object to a dictionary for JSON serialization
    def to_dict(self):
        return {'username': self.username, 'email': self.email, 'role': self.role}

    # Static method to create UserProfile object from a dictionary (for JSON deserialization)
    @staticmethod
    def from_dict(data):
        return UserProfile(data['username'], data['email'], data.get('role', 'user'))

def save_user_profile_json(profile, filename='user_profile_secure.json'):
    """Saves a user profile using JSON."""
    with open(filename, 'w') as f:
        json.dump(profile.to_dict(), f, indent=4) # Use indent for pretty-printing
    print(f"[JSON] Saved profile for {profile.username} to {filename}")

def load_user_profile_json(filename='user_profile_secure.json'):
    """Loads a user profile using JSON."""
    if not os.path.exists(filename):
        print(f"[JSON] File {filename} not found.")
        return None
    print(f"[JSON] Loading profile from {filename}...")
    with open(filename, 'r') as f:
        data = json.load(f)
    return UserProfile.from_dict(data)

# --- Demonstration ---
user2 = UserProfile('bob', 'bob@example.com', 'admin')
save_user_profile_json(user2)

loaded_user2 = load_user_profile_json()
if loaded_user2:
    print(f"[JSON] Loaded: {loaded_user2}")


# --- Simulating a malicious JSON payload ---
# If an attacker injects executable code into JSON, it will be treated as a string.
malicious_json_data = {
    "username": "hacker",
    "email": "hacker@example.com",
    "role": "user",
    "malicious_payload": "<script>alert('XSS from JSON!')</script>",
    "command_to_execute": "os.system('rm -rf /')" # This will just be a string in JSON
}

malicious_json_file = 'malicious_data.json'
with open(malicious_json_file, 'w') as f:
    json.dump(malicious_json_data, f, indent=4)

print(f"\n[JSON] Attempting to load malicious JSON data from {malicious_json_file}...")
try:
    malicious_json_object = load_user_profile_json(malicious_json_file)
    # The extra fields like 'malicious_payload' would be ignored by from_dict or just loaded as data.
    print(f"[JSON] Loaded (malicious data treated as data): {malicious_json_object}")
    # Verify that the malicious fields are just data
    if 'malicious_payload' in malicious_json_data:
        print(f"[JSON] Malicious payload detected as data: {malicious_json_data['malicious_payload']}")
    if 'command_to_execute' in malicious_json_data:
        print(f"[JSON] Command to execute detected as data: {malicious_json_data['command_to_execute']}")

except Exception as e:
    print(f"[JSON] Error loading malicious JSON: {e}")
finally:
    # Clean up generated files
    if os.path.exists('user_profile_secure.json'):
        os.remove('user_profile_secure.json')
    if os.path.exists(malicious_json_file):
        os.remove(malicious_json_file)
```
