Task 1 – Python to Java Conversion

Task:

Convert a Python program that performs file handling operations into an
equivalent Java program using AI
assistance.

Instructions:

• Prompt AI to translate a Python script
that reads from and writes
to a file into Java.

• Ensure proper handling of file streams, exceptions, and
input/output formatting.

• Verify that the Java program produces the same file content and
console output as the Python version.

• Test both implementations with sample input files.

Expected Output:
A correctly translated and executable Java program that performs the
same file handling operations as the original Python script.


Prompt:

Convert a Python file handling program (read/write operations) into an equivalent Java program.

Ensure proper use of file streams, exception handling, and input/output formatting in Java.

Maintain identical functionality, file content, and console output as the Python version.

Provide complete, executable Java code with sample input and output for testing.


```python
import os

# Python script for file handling
def python_file_handler():
    input_filename = 'input.txt'
    output_filename = 'output.txt'

    # Create a sample input file (if it doesn't exist)
    if not os.path.exists(input_filename):
        with open(input_filename, 'w') as f:
            f.write('Hello Python!\n')
            f.write('This is a test.\n')
            f.write('File handling example.')
        print('Python: input.txt created for demonstration.')
    else:
        print('Python: input.txt already exists.')

    try:
        # Read from input.txt and write modified content to output.txt
        with open(input_filename, 'r') as infile:
            with open(output_filename, 'w') as outfile:
                print(f'\nPython: Reading from {input_filename} and writing to {output_filename}')
                for line_num, line in enumerate(infile):
                    processed_line = line.strip().upper() + f' (processed by Python line {line_num + 1})\n'
                    outfile.write(processed_line)
                    print(f'Python: Processed line: {line.strip()} -> {processed_line.strip()}')
        print('Python: File handling completed successfully.')

        # Verify output.txt content
        print('\nPython: Content of output.txt:')
        with open(output_filename, 'r') as f:
            print(f.read())

    except FileNotFoundError:
        print(f'Python Error: The file {input_filename} was not found.')
    except Exception as e:
        print(f'Python Error: An unexpected error occurred: {e}')

# Run the Python file handler
python_file_handler()
```

Task 2 – C++ to Python Function Conversion

Translate a C++ program that implements a sorting algorithm into
Python using AI assistance.

Instructions:
• Provide AI with a C++ implementation of a sorting technique
(e.g., Bubble Sort or Selection Sort).

• Ask AI to generate an equivalent Python implementation.

• Ensure logical correctness and validate results using multiple
test cases.

• Confirm that the Python version maintains the same sorting
behavior.

Expected Output:
A correct and executable Python program equivalent to the given C++
sorting implementation


Prompt:

Convert the following C++ sorting program (e.g., Bubble Sort or Selection Sort) into an equivalent Python program.
Ensure the same logic and sorting behavior are preserved in the Python version.

Verify correctness by testing with multiple input cases and comparing outputs.

Provide clean, executable Python code with sample inputs and outputs.


```python
# Python implementation of Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def run_tests():
    print("Python: Running Bubble Sort tests")

    # Test Case 1
    arr1 = [64, 34, 25, 12, 22, 11, 90]
    print(f"\nOriginal array 1 (Python): {arr1}")
    bubble_sort(arr1)
    print(f"Sorted array 1 (Python): {arr1}")
    expected1 = [11, 12, 22, 25, 34, 64, 90]
    assert arr1 == expected1, f"Test Case 1 Failed: Expected {expected1}, got {arr1}"
    print("Test Case 1 Passed.")

    # Test Case 2
    arr2 = [5, 1, 4, 2, 8]
    print(f"\nOriginal array 2 (Python): {arr2}")
    bubble_sort(arr2)
    print(f"Sorted array 2 (Python): {arr2}")
    expected2 = [1, 2, 4, 5, 8]
    assert arr2 == expected2, f"Test Case 2 Failed: Expected {expected2}, got {arr2}"
    print("Test Case 2 Passed.")

    # Test Case 3: Already sorted array
    arr3 = [1, 2, 3, 4, 5]
    print(f"\nOriginal array 3 (Python): {arr3}")
    bubble_sort(arr3)
    print(f"Sorted array 3 (Python): {arr3}")
    expected3 = [1, 2, 3, 4, 5]
    assert arr3 == expected3, f"Test Case 3 Failed: Expected {expected3}, got {arr3}"
    print("Test Case 3 Passed.")

    # Test Case 4: Reverse sorted array
    arr4 = [5, 4, 3, 2, 1]
    print(f"\nOriginal array 4 (Python): {arr4}")
    bubble_sort(arr4)
    print(f"Sorted array 4 (Python): {arr4}")
    expected4 = [1, 2, 3, 4, 5]
    assert arr4 == expected4, f"Test Case 4 Failed: Expected {expected4}, got {arr4}"
    print("Test Case 4 Passed.")

    # Test Case 5: Empty array
    arr5 = []
    print(f"\nOriginal array 5 (Python): {arr5}")
    bubble_sort(arr5)
    print(f"Sorted array 5 (Python): {arr5}")
    expected5 = []
    assert arr5 == expected5, f"Test Case 5 Failed: Expected {expected5}, got {arr5}"
    print("Test Case 5 Passed.")

    # Test Case 6: Single element array
    arr6 = [7]
    print(f"\nOriginal array 6 (Python): {arr6}")
    bubble_sort(arr6)
    print(f"Sorted array 6 (Python): {arr6}")
    expected6 = [7]
    assert arr6 == expected6, f"Test Case 6 Failed: Expected {expected6}, got {arr6}"
    print("Test Case 6 Passed.")

    print("\nAll Python Bubble Sort tests completed.")

run_tests()
```

Task 3 – SQL to Pandas Query

Convert a SQL query involving GROUP BY and aggregate functions
into an equivalent Pandas Data Frame operation.

Instructions:

• Provide AI with a SQL query that calculates aggregate values
(e.g., total sales per department).

• Prompt AI to generate equivalent Pandas code using groupby()
and aggregation functions.

• Test the generated code using a sample dataset.

• Verify that both SQL and Pandas outputs match.
Expected Output:

An accurate Pandas Data Frame operation that produces results
equivalent to the given SQL query with aggregation.

Prompt:

Convert the following SQL query into equivalent Pandas code using groupby() and aggregation functions.

SQL Query:
SELECT department, SUM(sales) AS total_sales, AVG(sales) AS avg_sales
FROM sales_data
GROUP BY department;

Also provide a sample dataset and ensure the Pandas output matches the SQL result.

```python
import pandas as pd

# 1. Create a sample dataset (simulating sales_data table)
data = {
    'department': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Clothing', 'Electronics', 'Books', 'Clothing', 'Books'],
    'sales': [1200, 25, 50, 15, 60, 75, 40, 85, 100]
}
sales_df = pd.DataFrame(data)
print("--- Original Sales Data (Pandas DataFrame) ---")
display(sales_df)

# 2. Convert the SQL query to equivalent Pandas code
#    GROUP BY 'department' and calculate SUM(sales) and AVG(sales)
pandas_result = sales_df.groupby('department').agg(
    total_sales=('sales', 'sum'),
    avg_sales=('sales', 'mean')
).reset_index()

print("\n--- Pandas Result (Equivalent to SQL Query) ---")
display(pandas_result)

# 3. Verification: Manually calculate expected result and compare
expected_data = {
    'department': ['Books', 'Clothing', 'Electronics'],
    'total_sales': [155, 170, 1325],
    'avg_sales': [51.666667, 56.666667, 441.666667]
}
expected_df = pd.DataFrame(expected_data)

# Round the 'avg_sales' in expected_df and pandas_result for comparison due to floating point precision
pandas_result['avg_sales'] = pandas_result['avg_sales'].round(6)
expected_df['avg_sales'] = expected_df['avg_sales'].round(6)

print("\n--- Expected Result (for verification) ---")
display(expected_df)

# Sort both for reliable comparison
pandas_result_sorted = pandas_result.sort_values(by='department').reset_index(drop=True)
expected_df_sorted = expected_df.sort_values(by='department').reset_index(drop=True)

is_equal = pandas_result_sorted.equals(expected_df_sorted)

print(f"\nVerification: Do Pandas result and Expected result match? {is_equal}")
if is_equal:
    print("The Pandas output perfectly matches the expected result.")
else:
    print("There is a mismatch between the Pandas output and the expected result.")
```

### Task 4: Multi-Language Snippet Converter - Python to Java Factorial Function

**Scenario:** We will convert a basic mathematical function (factorial calculation) from Python to Java to illustrate language translation while preserving functionality.

**Python Source Code:**

```python
# Python implementation of a factorial function
def factorial(n):
    # Handle negative numbers
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    # Base case for factorial of 0
    if n == 0:
        return 1
    # Calculate factorial iteratively
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

print("--- Python Factorial Function Tests ---")
test_numbers = [0, 1, 5, 7, 10]
for num in test_numbers:
    try:
        print(f"Factorial of {num} is {factorial(num)}")
    except ValueError as e:
        print(f"Error for {num}: {e}")

# Test with a negative number to check error handling
try:
    print(f"Factorial of -3 is {factorial(-3)}")
except ValueError as e:
    print(f"Error for -3: {e}")
```
