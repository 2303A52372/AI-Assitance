```python
class TodoListManager:
    def __init__(self):
        self.tasks = []
        self.next_id = 1

    def add_task(self, description):
        task = {
            'id': self.next_id,
            'description': description,
            'completed': False
        }
        self.tasks.append(task)
        self.next_id += 1
        print(f"Task '{description}' added.")

    def view_tasks(self):
        if not self.tasks:
            print("No tasks in the list.")
            return
        print("\n--- To-Do List ---")
        for task in self.tasks:
            status = "[X]" if task['completed'] else "[ ]"
            print(f"{task['id']}. {status} {task['description']}")
        print("------------------")

    def mark_task_complete(self, task_id):
        for task in self.tasks:
            if task['id'] == task_id:
                task['completed'] = True
                print(f"Task {task_id} marked as complete.")
                return
        print(f"Task with ID {task_id} not found.")

    def run(self):
        while True:
            print("\n--- Menu ---")
            print("1. Add Task")
            print("2. View Tasks")
            print("3. Mark Task Complete")
            print("4. Exit")
            choice = input("Enter your choice: ")

            if choice == '1':
                description = input("Enter task description: ")
                self.add_task(description)
            elif choice == '2':
                self.view_tasks()
            elif choice == '3':
                try:
                    task_id = int(input("Enter the ID of the task to mark complete: "))
                    self.mark_task_complete(task_id)
                except ValueError:
                    print("Invalid input. Please enter a number.")
            elif choice == '4':
                print("Exiting To-Do List Manager. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == '__main__':
    manager = TodoListManager()
    # You can uncomment the line below to run the manager directly.
    # manager.run()
    print("To run the To-Do List Manager, execute `manager.run()` in a new cell after this cell.")
```

```python
from datetime import datetime

class TodoListManager:
    def __init__(self):
        self.tasks = []
        self.next_id = 1

    def add_task(self, description, deadline=None, priority='medium'):
        task = {
            'id': self.next_id,
            'description': description,
            'completed': False,
            'deadline': deadline,
            'priority': priority.lower() # low, medium, high
        }
        self.tasks.append(task)
        self.next_id += 1
        print(f"Task '{description}' added (Priority: {priority}, Deadline: {deadline or 'None'}).")

    def view_tasks(self, sort_by=None, reverse=False):
        if not self.tasks:
            print("No tasks in the list.")
            return

        display_tasks = list(self.tasks)

        if sort_by == 'priority':
            priority_order = {'high': 1, 'medium': 2, 'low': 3}
            display_tasks.sort(key=lambda x: priority_order.get(x['priority'], 99), reverse=reverse)
        elif sort_by == 'deadline':
            display_tasks.sort(key=lambda x: datetime.strptime(x['deadline'], '%Y-%m-%d') if x['deadline'] else datetime.max, reverse=reverse)

        print("\n--- To-Do List ---")
        for task in display_tasks:
            status = "[X]" if task['completed'] else "[ ]"
            deadline_str = f" (Due: {task['deadline']})" if task['deadline'] else ""
            priority_str = f" (Priority: {task['priority'].capitalize()})"
            print(f"{task['id']}. {status} {task['description']}{priority_str}{deadline_str}")
        print("------------------")

    def mark_task_complete(self, task_id):
        for task in self.tasks:
            if task['id'] == task_id:
                task['completed'] = True
                print(f"Task {task_id} marked as complete.")
                return
        print(f"Task with ID {task_id} not found.")

    def search_tasks(self, keyword):
        results = [task for task in self.tasks if keyword.lower() in task['description'].lower()]
        if not results:
            print(f"No tasks found containing '{keyword}'.")
            return
        print(f"\n--- Search Results for '{keyword}' ---")
        for task in results:
            status = "[X]" if task['completed'] else "[ ]"
            deadline_str = f" (Due: {task['deadline']})" if task['deadline'] else ""
            priority_str = f" (Priority: {task['priority'].capitalize()})"
            print(f"{task['id']}. {status} {task['description']}{priority_str}{deadline_str}")
        print("------------------------------------")

    def run(self):
        while True:
            print("\n--- Enhanced To-Do Menu ---")
            print("1. Add Task")
            print("2. View Tasks")
            print("3. Mark Task Complete")
            print("4. Search Tasks")
            print("5. Exit")
            choice = input("Enter your choice: ")

            if choice == '1':
                description = input("Enter task description: ")
                deadline_input = input("Enter deadline (YYYY-MM-DD, leave blank if none): ")
                deadline = deadline_input if deadline_input else None
                priority = input("Enter priority (low, medium, high, default: medium): ") or 'medium'
                self.add_task(description, deadline, priority)
            elif choice == '2':
                sort_option = input("Sort by (priority/deadline/none, default: none): ")
                reverse_sort = input("Reverse order? (yes/no, default: no): ").lower() == 'yes'
                self.view_tasks(sort_by=sort_option if sort_option != 'none' else None, reverse=reverse_sort)
            elif choice == '3':
                try:
                    task_id = int(input("Enter the ID of the task to mark complete: "))
                    self.mark_task_complete(task_id)
                except ValueError:
                    print("Invalid input. Please enter a number.")
            elif choice == '4':
                keyword = input("Enter keyword to search: ")
                self.search_tasks(keyword)
            elif choice == '5':
                print("Exiting Enhanced To-Do List Manager. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == '__main__':
    enhanced_manager = TodoListManager()
    print("To run the Enhanced To-Do List Manager, execute `enhanced_manager.run()` in a new cell after this cell.")
```

```python
from datetime import datetime

class TodoListManager:
    def __init__(self):
        self.tasks = []
        self.next_id = 1

    def add_task(self, description, deadline=None, priority='medium'):
        task = {
            'id': self.next_id,
            'description': description,
            'completed': False,
            'deadline': deadline,
            'priority': priority.lower() # low, medium, high
        }
        self.tasks.append(task)
        self.next_id += 1
        print(f"Task '{description}' added (Priority: {priority}, Deadline: {deadline or 'None'}).")

    def view_tasks(self, sort_by=None, reverse=False):
        if not self.tasks:
            print("No tasks in the list.")
            return

        display_tasks = list(self.tasks)

        if sort_by == 'priority':
            priority_order = {'high': 1, 'medium': 2, 'low': 3}
            display_tasks.sort(key=lambda x: priority_order.get(x['priority'], 99), reverse=reverse)
        elif sort_by == 'deadline':
            display_tasks.sort(key=lambda x: datetime.strptime(x['deadline'], '%Y-%m-%d') if x['deadline'] else datetime.max, reverse=reverse)

        print("\n--- To-Do List ---")
        for task in display_tasks:
            status = "[X]" if task['completed'] else "[ ]"
            deadline_str = f" (Due: {task['deadline']})" if task['deadline'] else ""
            priority_str = f" (Priority: {task['priority'].capitalize()})"
            print(f"{task['id']}. {status} {task['description']}{priority_str}{deadline_str}")
        print("------------------")

    def mark_task_complete(self, task_id):
        for task in self.tasks:
            if task['id'] == task_id:
                task['completed'] = True
                print(f"Task {task_id} marked as complete.")
                return
        print(f"Task with ID {task_id} not found.")

    def search_tasks(self, keyword):
        results = [task for task in self.tasks if keyword.lower() in task['description'].lower()]
        if not results:
            print(f"No tasks found containing '{keyword}'.")
            return
        print(f"\n--- Search Results for '{keyword}' ---")
        for task in results:
            status = "[X]" if task['completed'] else "[ ]"
            deadline_str = f" (Due: {task['deadline']})" if task['deadline'] else ""
            priority_str = f" (Priority: {task['priority'].capitalize()})"
            print(f"{task['id']}. {status} {task['description']}{priority_str}{deadline_str}")
        print("-------------------------------------")

    def run(self):
        while True:
            print("\n--- Enhanced To-Do Menu ---")
            print("1. Add Task")
            print("2. View Tasks")
            print("3. Mark Task Complete")
            print("4. Search Tasks")
            print("5. Exit")
            choice = input("Enter your choice: ")

            if choice == '1':
                description = input("Enter task description: ")
                deadline_input = input("Enter deadline (YYYY-MM-DD, leave blank if none): ")
                deadline = deadline_input if deadline_input else None
                priority = input("Enter priority (low, medium, high, default: medium): ") or 'medium'
                self.add_task(description, deadline, priority)
            elif choice == '2':
                sort_option = input("Sort by (priority/deadline/none, default: none): ")
                reverse_sort = input("Reverse order? (yes/no, default: no): ").lower() == 'yes'
                self.view_tasks(sort_by=sort_option if sort_option != 'none' else None, reverse=reverse_sort)
            elif choice == '3':
                try:
                    task_id = int(input("Enter the ID of the task to mark complete: "))
                    self.mark_task_complete(task_id)
                except ValueError:
                    print("Invalid input. Please enter a number.")
            elif choice == '4':
                keyword = input("Enter keyword to search: ")
                self.search_tasks(keyword)
            elif choice == '5':
                print("Exiting Enhanced To-Do List Manager. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == '__main__':
    enhanced_manager = TodoListManager()
    print("To run the Enhanced To-Do List Manager, execute `enhanced_manager.run()` in a new cell after this cell.")
```
